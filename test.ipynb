```python
import requests
from bs4 import BeautifulSoup
```

### 查找四大指數

```python
url="https://histock.tw/%E5%9C%8B%E9%9A%9B%E8%82%A1%E5%B8%82"
```

```python
resp=requests.get(url)
resp
```

```python
resp.text
```

```python
soup=BeautifulSoup(resp.text,"lxml")
soup
```

```python
trs=soup.find(string="加權指數").find_parent("div").find_all("tr")
trs
```

```python
datas=[]
for tr in trs:
    data=[]
    for th in tr.find_all("th"):
        print(th.text.strip(),end="\t")
        data.append(th.text.strip())

    for td in tr.find_all("td"):
        print(td.text.strip(),end="\t")
        data.append(td.text.strip())

    datas.append(data)
    print()

datas
```

### 六督平均值

```python
def convert_value(value):
    try:
        return eval(value)
    except:
        return None
```

```python
df["pm25"] = df["pm25"].apply(convert_value)
# 移除有None的數據
df = df.dropna()
df
     
```

```python
df["county"].unique()
```

```python
six_countys=['臺北市','新北市','桃園市','臺中市','臺南市','高雄市']
six_countys
```

```python
df.groupby("county").get_group('臺北市')["pm25"].mean()
```

```python
for county in six_countys:
    avg_pm25=df.groupby("county").get_group(county)["pm25"].mean()
    print(county,avg_pm25)
```
